## Missing Values

```python
dataset.isnull().sum()
```

```python
features_with_na = [features for features in dataset.columns if dataset[features].isnull().sum()>1]
```

```python
features_with_na
```

```python
dataset['Electrical'].isnull().mean()
```

```python
for features in features_with_na:
    print(features , np.round(dataset[features].isnull().mean(),4), '% missing values')
```

```python
data = dataset.copy()
for feature in features_with_na:
    # Make a variable that display 1 if observation is missing :
    data[feature]=np.where(data[feature].isnull(),1,0)
    # Lets calculate the mean sales price where information is missing or present:
    data.groupby(feature)['SalePrice'].median().plot.bar()
    plt.title(feature)
    plt.show()
```

Here the relationship between missing values and the dependent variable is clearly visible , which will help us a lot in feature selection.

# Temporal Variables(eg. Datetime var.)

```python
year_fet = [feature for feature in num_fet if 'Yr' in feature or 'Year' in feature]
```

```python
year_fet
```

```python
for feature in year_fet:
    print(feature,dataset[feature].unique())
```

```python
dataset.groupby('YrSold')['SalePrice'].median().plot.line()
plt.xlabel("Year Sold")
plt.ylabel("Median House Price")
plt.title("House Price vs Year Sold")
```

```python
# Here we'll find out the difference between all year features and sales price:
```

```python
for feature in year_fet :
    if feature != 'YrSold':
        dataset[feature] = data['YrSold']-data[feature]
        plt.scatter(data[feature],data['SalePrice'])
        plt.xlabel('Garage Yr Built')
        plt.ylabel('Sale Price')
        plt.show()
```

Within all the yearwise information , it's clear that the new houses are more costlier than older houses , 

Pareto or a Power Law Distribution.

```python
'''
* Numerical Variables :
* Continious Var
* Discrete Var
'''
```

```python
disccrete_feature = [feature for feature in num_fet if len(dataset[feature].unique())<=25]
```

```python
len(disccrete_feature)
```

```python
dataset[disccrete_feature].head()
```

To find relationship between discrete features and SalesPrice

```python
for feature in disccrete_feature:
    dataset.groupby(feature)['SalePrice'].median().plot.bar()
    plt.xlabel("Discrete Feature")
    plt.ylabel("Sale Price")
    plt.show()
    plt.title(feature)
```

```python
#Continious Feature

continiou_fet = [feature for feature in num_fet if feature not in disccrete_feature+year_fet]
print(f"{len(continiou_fet)}  \n \n {continiou_fet}")
```

```python
# Lets analyse the cont. values in respect to sales price:
data=dataset.copy()
for feature in continiou_fet:
    dataset.groupby(feature)['SalePrice'].median().plot.hist()
    plt.xlabel("Discrete Feature")
    plt.ylabel("Sale Price")
    plt.show()
    plt.title(feature)
```

```python
data=dataset.copy()
for feature in continiou_fet:
    dataset[feature].hist(bins=25)
    plt.xlabel(feature)
    plt.ylabel("Sale Price")
    plt.show()
    plt.title(feature)
```

## Logarithmic Transformation:

```python
data = dataset.copy()
for i in continiou_fet:
    if 0 in data[i].unique():
        pass
    else :
        data[i] = np.log(data[i])
        plt.scatter(data[i],data["SalePrice"])
        plt.xlabel(i)
        plt.ylabel("Sale Price")
        plt.title(i)
        plt.show()
```

# OUTLIERS

```python
data = dataset.copy()
for i in continiou_fet:
    if 0 in data[i].unique():
        pass
    else :
        data[i] = np.log(data[i])
        data.boxplot(column = i)
        plt.xlabel(i)
        plt.ylabel(i)
        plt.title(i)
        plt.show()
```
